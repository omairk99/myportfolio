```python
data = [[], [], []]
for i in range(3):
    data[i]= np.random.randn(100).cumsum()
    
```

```python
styles = ['ko-', 'gs--', 'b.-.']
labels = ['one', 'two','three']
```

```python
for i in range(3):
    plt.plot(data[i], styles[i], label = labels[i])
plt.legend(loc = 'best')
# plt.text(10, 5,'Just to say hi', family= 'monospace', fontsize = 10)
y = data[0].max()
x = data[0].argmax()

plt.annotate('Peak One', xy = (x+0.1, y+0.1), xytext = (x+5, y+5), arrowprops = dict(facecolor = 'black', headwidth = 4, width= 2, headlength = 4),
                                                                                    horizontalalignment = 'left', verticalalignment = 'top')

filename = 'updown'
plt.savefig(filename)
```
